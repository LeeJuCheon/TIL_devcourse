```python
x = np.arange(-10,10,0.01)
plt.plot(x,x**2)
plt.show()
```

```python
x = np.arange(-10,10,0.01)
plt.title("y= x^2 graph")
plt.xlabel("x value")
plt.ylabel("f(x) value")

plt.axis([-5,5,0,25])

plt.xticks([i for i in range(-5,5,1)])
plt.yticks([i for i in range(0,27,3)])

plt.plot(x,x**2,label ="trend")
plt.legend()
plt.show()
```

```python
# 꺾은선 그래프 (plot)

x= np.arange(20)   #0~ 19
y = np.random.randint(0,20,20)

plt.plot(x,y)
plt.axis([0,20,0,20])
plt.yticks([i for i in range(0,25,5)])
plt.show()
```

```python
# scatter plot
plt.scatter(x,y)
plt.show()
```

```python
#boxplot
plt.boxplot((x,y))
plt.title("Box plot of x, y")
plt.show()
```

```python
# barplot
plt.bar(x,y)
plt.xticks([i for i in range(0,20,1)])
plt.show()
```

```python
#히스토그램 hist
plt.hist(y, bins=np.arange(0,20,2))
plt.xticks(np.arange(0,20,2))
plt.show
```

```python
# 원형그래프 Pie chart
z=[100,300,200,400]

plt.pie(z,labels=['one','two','three','four'])

plt.show()
```

```python
# Cat plot
# 데이터셋을 위한 sklearn import
from sklearn import datasets

dataset = datasets.load_iris()
test_df = pd.DataFrame(dataset.data,columns=dataset.feature_names)
test_df['Classes']=dataset.target
target=["setosa", "versicolor", "virginica"]
for i in range(3):
    test_df=test_df.replace({'Classes':i},target[i])
```

```python
test_df.head()
```

```python
s=sns.catplot(x='Classes',y='sepal length (cm)', data=test_df)
s.fig.set_size_inches(10,6)
plt.show()
```

```python
# 스트립 플롯
s= sns.stripplot(x='Classes',y='sepal length (cm)', data=test_df)
```

```python
# 히트맵 
test_df2 = pd.DataFrame(dataset.data,columns=dataset.feature_names)

sns.heatmap(test_df2.corr())
plt.show()
```
